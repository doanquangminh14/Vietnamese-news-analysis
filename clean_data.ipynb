```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def draw_wordcloud(dataset):
    print("Đang vẽ Đám mây từ vựng...")
    all_words = " ".join(dataset['Text_Segmented'].astype(str))
    
    wordcloud = WordCloud(
        width=1000, 
        height=500, 
        background_color='white', 
        colormap='tab20',
        max_words=100, 
        collocations=False 
    ).generate(all_words)
    
    
    plt.figure(figsize=(12, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Top 100 Từ Khóa Xuất Hiện Nhiều Nhất', fontsize=20, fontweight='bold', pad=20)
    plt.show()

if __name__ == "__main__":
    
    df_clean = pd.read_csv("vnexpress_data_clean.csv")
    
   
    draw_wordcloud(df_clean)
```
