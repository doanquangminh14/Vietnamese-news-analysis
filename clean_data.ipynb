```python
df1 = pd.read_csv('data_output/vnexpress_ai_insight.csv')
df1
```

```python
df1.isnull().sum()
```

```python
df1 = df1.rename(columns={'AI_Insight':'Label'})  # hoặc  df1.rename(columns={'AI_Insight':'Label'}, inplace=True)
df1
```

```python
df1['Label'] = df1['Label'].astype(str).str.strip()
df_clean = df1[df1['Label'] != "Lỗi"].dropna(subset=['Label'])
```

hết token lỗi  1257->

```python
df_final = df_clean.iloc[:1257].copy()
```

```python
df_final['Label'].value_counts().plot(kind='bar')
```

```python
df_final['Label'] = df_final['Label'].str.replace('^- ', '', regex=True).str.strip().str.capitalize()

mapping = {
    'Thông tin': 'Thông tin',
    '- thông tin': 'Thông tin',
    '- cảnh báo': 'Cảnh báo',
    '- không': 'Không'
}
df_final['Label'] = df_final['Label'].replace(mapping)
```

```python
df_final['Label'].value_counts().plot(kind='bar')
```

```python
df_final.to_csv("data_output/vnexpress_final_results.csv", index=False, encoding='utf-8-sig')
```
